```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xlwt
import math
import random

from xlwt import Workbook

#------Değiştirilecek yerler------

Starting_WaveLength = 380   #Başlangıç Dalga Boyu Uzunluk Değeri
Ending_WaveLength = 780     #Bitiş Dalga Boyu Uzunluk Değeri
DalgaBoyları_Fark = 5       #Ardışık iki Dalga Arasındaki Sabit Farki Örneğin 380 385 390 veya 380 381 382... gibi

DataSayısı = 11             #input Excel Tabloasındaki kullanacağınız Data Sayınız

#Kullanmak istediğiniz Dataların isimlerinin girildiği yer.Çok uzun olmamalı.DataSayısı ile aynı sayıda isim girilmeli.
DatasNames = ["FiK","FiR","Rr","Sxi","Sai","Sbi","Sei","Syi","Ski","Sni","Ssi"]

formul_no = 2 #Hangi formul denklemi ve input Excel dosyası kullanılacaksa ona göre girilmeli.Örneğin input_1 input_2 gibi

Iteration_Number = 1000  #Monte Carlo İterasyon Sayısı

#---------------------------------




Sum_WaveLength = int((Ending_WaveLength - Starting_WaveLength+DalgaBoyları_Fark)/DalgaBoyları_Fark)
print("Dalga Boyu Sayısı:"+str(Sum_WaveLength))

Input_Datas_Values=[] #input Excel tablosundan verilerin çekilip atıldığı Liste
Datas_List=[] #Dataların random üretelip her bir datanın Verilerinin atıldığı List.Liste içinde Liste barındırmaktadır.
for i in range(0,DataSayısı):
    list = []
    Datas_List.append(list)

sum_of_sample_List=[] #for Standart Dev calculate

#Bunları tek bir fonksiyonda yapmalısın.Fazladan satır olmasın
for i in range(0,Sum_WaveLength):
    liste = []
    sum_of_sample_List.append(liste)

wb = Workbook() #Başlangıç output excel dosyalarının oluşturulması
tiles = wb.add_sheet('output2')
tiles2 = wb.add_sheet('output1')
tiles2.write(0,0,'Iteration Number') 
tiles2.write(0,1,Iteration_Number)

from openpyxl import Workbook,load_workbook #En tepede tanımlandığında error verdiği için arada tanımladım.
wB= load_workbook("input_"+str(formul_no)+".xlsx")#Uzantısının xlsx olmalı.input dosyası okunur ve wS Workbook'una atılır
wS = wB.active
wS = wB["INPUT"]
ColumnNames = ['Wave Length','Average', 'Standart Deviation']
for col_num,data in enumerate(ColumnNames):
    tiles2.write(1,col_num,data)

#Bu değişkenlerde ise DataSayımız kadar datamızın ortalama değerlerini ve standart sapmalarını yazıcaz.
#Tek olanlar ortalama çift olanlar standart sapmayı veriyor.   
AVG_STD_of_DATA=[]
AVG_STD_of_Result=[]

KorelasyonaMatrix = []
TempMatrix = [] #Gerekli Matrisi kullanmak için kullanacağız
for i in range(0,Sum_WaveLength):
    List = []
    KorelasyonaMatrix.append(List)
    for j in range(0,DataSayısı):
        SutunList = []
        KorelasyonaMatrix[i].append(SutunList)
        
    TempSutunList = []
    TempMatrix.append(TempSutunList)

print("Temp Matrix:\n"+str(TempMatrix))        
        
def Data_Find_AVG(Data):
    
    ortalama=np.mean(Data)
    return ortalama

def Data_Find_STD(Data):
   
    standartS=np.std(Data)
    return standartS
def get_rand_values(avg,std):
    
    result=np.random.normal(avg,abs(std),1)#Belki bu da örnek sayısına göre değişebilir.Zamanda oynama olarak.
    ortalama=np.mean(result)
    return ortalama

def formul(DL):
    
    #Buradaki datalarımızın sırasının ne olduğu DatasNames sırasındaki data isimleri gibi yazılmalı.
    #Sonradan eklenilmesi gereken formüller yeni bir elif altında eklenilmeli ve oluşturulmalı.
    
    if formul_no == 1:                                                                                                                                                                                
        result = (DL[0]+DL[5])*(DL[4]/DL[1])*(DL[2]/DL[3])*(1+DL[6]+DL[7]+DL[12]+DL[8]+DL[9]+DL[10]+DL[11])
    elif formul_no == 2:
        result =(DL[0]/DL[1])*DL[4]*(DL[2]*DL[3]*DL[5]*DL[6]*DL[7]*DL[8]*DL[9]*DL[10])
    
    return result

#output2   
def CreateExcel():#Create Excel Titles and Write First Values of Datas
    
    ColumnNames = ['Wave Length']
    for i in range(0,DataSayısı):
        ColumnNames.append('Average '+DatasNames[i])
        ColumnNames.append('Standart Dev '+DatasNames[i])
    
    for col_num,data in enumerate(ColumnNames):
        tiles.write(0,col_num,data)
        
    tiles.write(1,0,Starting_WaveLength)
    
    for col_num, data in enumerate(AVG_STD_of_DATA):
        tiles.write(1, col_num+1, data)
    
    wb.save('OUTPUT_'+str(Iteration_Number)+'_'+str(formul_no)+'.xls')

#output2      
def AddToExcel(DalgaBoyuIndexLast,count):#add to output excel 
    
    tiles.write(count,0,DalgaBoyuIndexLast)
    
    for col_num,data in enumerate(AVG_STD_of_DATA):
        tiles.write(count,col_num+1,data)
    
    wb.save('OUTPUT_'+str(Iteration_Number)+'_'+str(formul_no)+'.xls')
    
#output1    
def Output1Excel(DalgaBoyuIndex,count,avg,std):
    
    List=[avg,std]
    tiles2.write(count+1,0,DalgaBoyuIndex)
    for col_num,data in enumerate(List):
        tiles2.write(count+1,col_num+1,data)
        
    wb.save('OUTPUT_'+str(Iteration_Number)+'_'+str(formul_no)+'.xls')  

def DefineCorrMatrix(Matrix):
    #Array tipinde Matrix olmalı.Frame olarak değil.
    temp = np.corrcoef(Matrix)
    return np.array(temp)    

def corrPlot(Matrix):

    fig=plt.figure()
    subplot=fig.add_subplot(111)
    cax=subplot.imshow(Matrix,vmin=-1,vmax=1,interpolation ='nearest',cmap="jet",origin="lower")
    fig.colorbar (cax, ticks=[-1,-0.5,0,0.5,1])
    plt.show()

def monte_carlo(IterationNum,DalgaBoyuIndexLast,count):
    
    #count yazılacağı index değerine karşılık geliyor.
    
    sum_of_samples=0.0
    
    for i in range(IterationNum):
        
        Indexsayac=0
        mean_data=[]
        
        for i in range(0,DataSayısı):#random dataların oluşturulup atandığı yer.
            mean_data.append(get_rand_values(Input_Datas_Values[Indexsayac],Input_Datas_Values[Indexsayac+1]))
            KorelasyonaMatrix[count-1][i].append(mean_data[i])
            Datas_List[i].append(mean_data[i]) 
            Indexsayac+=2
        
        sum_of_sample_List[count-1].append(formul(mean_data)) #main döngsünde kullanılıyor.
        
        sum_of_samples+=formul(mean_data)
        
    print("------")
    print(sum_of_samples)
    
    #Bu bu değişkenlerde ise monte-carlo fonksiyonun içerisinde yaparak ortalama ve std buluyoruz.
    for k in range(0,DataSayısı):
        AVG_STD_of_DATA.append(Data_Find_AVG(Datas_List[k]))
        AVG_STD_of_DATA.append(Data_Find_STD(Datas_List[k]))
    
    if count == 1: #İlk kez döngüde dönüyorsa output2 Exceli oluştur.
        CreateExcel()
    else:
        AddToExcel(DalgaBoyuIndexLast,count)
        
    return float(sum_of_samples/IterationNum)

#-----MAIN-----

#Starting_WaveLength'in Excel de olup olmadığına bakılır ve hangi excel'deki hangi satıra denk geldiğini bulunur.
StartingRow = 1
count = 1
for satır in range(2,(Sum_WaveLength+1)*10):#+1 yapma sebebim Dalga Boyu Sayısının 1 fazla excel'de satıra denk gelmesi.
    if wS.cell(satır,1).value == Starting_WaveLength:
        print("StartingRow değişmektedir.")
        StartingRow = satır
        break
print("\nBaşlangıç Satır Indexi:"+str(StartingRow)+"\n")

for DalgaBoyuIndex in range(Starting_WaveLength,Ending_WaveLength+DalgaBoyları_Fark,DalgaBoyları_Fark):
    
    print("Dalga Boyu")
    print(DalgaBoyuIndex)
    print("count")
    print(count)
    
    #Dosyadan okuma işlemi.input dosyasından hücre hücre okuma işlemi
    for sutun in range(2,DataSayısı*2+2):#DataSayısı*2+2 de çalışıyor döngü tam olarak.2'den başladığı için
        Input_Datas_Values.append(float(wS.cell(StartingRow,sutun).value))    
    print("\nInput Data Values:\n"+str(Input_Datas_Values))
    
    Monte_Avg=monte_carlo(Iteration_Number,DalgaBoyuIndex,count)
    AVG_STD_of_Result.append(Monte_Avg)
    AVG_STD_of_Result.append(Data_Find_STD(sum_of_sample_List[count-1]))
    Output1Excel(DalgaBoyuIndex,count,AVG_STD_of_Result[len(AVG_STD_of_Result)-2],AVG_STD_of_Result[len(AVG_STD_of_Result)-1])
         
    for i in range(0,DataSayısı):
        Datas_List[i].clear()
    
    print("******")
    print(AVG_STD_of_DATA)
    AVG_STD_of_DATA.clear()
    Input_Datas_Values.clear()
    count+=1
    StartingRow+=1
    
print("Işık Akısının ortalama değeri:")
print(Monte_Avg)
print("*****")
print(AVG_STD_of_DATA)

#Korelasyon için diğer kısımlar

KorelasyonaArray = np.array(KorelasyonaMatrix)
print("KorelasyonaArray:"+str(KorelasyonaArray.shape))

for i in range(0,DataSayısı):
    for k in range(0,Sum_WaveLength):
        for j in range(0,Iteration_Number):
            TempMatrix[k].append(KorelasyonaMatrix[k][i][j])
            
    TempDizi = np.array(TempMatrix)
    print("TempMatrix:\n"+str(TempDizi))       
    print("Boyut:"+str(TempDizi.shape))
    
    print("\n*******\n "+str(i+1)+".data'nın Korelasyon Kat sayısı ve görselleştirilmesi:\n")
    CorrTemp = DefineCorrMatrix(TempDizi)
    print("Korelasyon Kat Sayı Matrisi:\n"+str(CorrTemp))
    print("Boyut:"+str(CorrTemp.shape))
   
    corrPlot(CorrTemp)
    
    #Burada Korelasyon İşlemini ve Görselleştirmesini yapman gerekiyor.Bunun üstünde
    TempMatrix.clear()
    for satir in range(0,Sum_WaveLength):#Bunu bir fonksiyon içerisinde yazarsan senin için daha iyi olur.Diğerlerinide ona göre yaparsın.
        TempSutunList = []
        TempMatrix.append(TempSutunList)
        
KorelasyonSonuçlar_Matrix = np.array(sum_of_sample_List)
print("Sonuçlar Matrix:\n"+str(KorelasyonSonuçlar_Matrix))
print("Boyut:"+str(KorelasyonSonuçlar_Matrix.shape))

Result_Corr = DefineCorrMatrix(KorelasyonSonuçlar_Matrix)
print("\nSonuçlar Korelasyon Kat Sayılar Matrisi:\n"+str(Result_Corr))
print("Boyut:"+str(Result_Corr.shape))

corrPlot(Result_Corr)

#--------------
```
